#### Question: Account managers would like to get in touch with as many of their clients as possible each month. In which month of the year does the team contact the greatest percentage of its clients?

```python
data.info()
```

```python
print("# of unique Date of Contact:", data['Date of Contact'].unique().size)
```

```python
def convert_datetime(row):
    d = datetime.strptime(str(row['Date of Contact']),'%Y-%m-%d %H:%M:%S')
    return str(d.year) + '-' + str(d.month)
```

```python
data['year_month'] = data.apply(convert_datetime,axis=1)
```

```python
data.head(5)
```

```python
contact_count = pd.crosstab(data.year_month, data['Account manager'])
```

```python
contact_count.sum().sum()
```

```python
contact_count['Total_count_percentage'] = contact_count.sum(axis=1)/1000
```

```python
contact_count.sort_values(by='Total_count_percentage')
```

#### Answer: In October 2013, the team contact the greatest percentage of its clients, which is 6% of the total contact number.
